```python
# Team', 'NOC', 'Games', 'Year', 'City', 'Sport', 'Event', 'Medal
def medal(df,year,country):
    medal = df.drop_duplicates(['Team', 'NOC', 'Games', 'Year', 'City', 'Sport', 'Event', 'Medal'])
    flag = 0

    if year=='Overall' and country=='Overall':
        tempdf = medal
    if year != 'Overall' and country == 'Overall':
        tempdf = medal[medal['Year']== year]
    if year == 'Overall', and country != 'Overall':
        tempdf = medal[medal['region'] == country]
    if year != 'Overall' and country !='Overall':
        tempdf = medal[(medal['region'] == country) and (medal['year'] == year)]
    
    if flag==1:
        x = medal.groupby('Year').sum()[['Bronze','Gold','Silver']].reset_index()
    else:
        x = medal.groupby('region').sum()[['Bronze','Gold','Silver']].reset_index()       
    return x
```
